```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as ply
import seaborn as sns
```

```python
df= pd.read_csv('data_to_be_cleaned.csv')
```

```python
df.head(30)
```

```python
df.info()
```

```python
new_df= df.copy()
```

```python
new_df.dropna(subset=['Client'], inplace=True)
new_df
```

```python
new_df.drop_duplicates(subset=['Client','Beneficiary','AmountUSD','TransactionDateTime'], inplace=True)
new_df
```

```python
new_df['Client'].str.replace("[a-zA-Z0-9 ]","", regex=True).unique()
```

```python
new_df['Client']=new_df['Client'].str.strip("<>*.- ")
new_df['Client']=new_df['Client'].str.replace('*',' ')
new_df
```

```python
new_df['Client']=new_df['Client'].str.upper()
new_df['Beneficiary']=new_df['Beneficiary'].str.upper()
new_df
```

```python
mean_value= new_df['AmountUSD'].mean()
median_value= new_df['AmountUSD'].median()

ax=new_df['AmountUSD'].plot(kind='hist', bins=10, color= 'skyblue', edgecolor= 'black')
ax.set_xlabel('Amount')

ax.axvline(mean_value, color='pink', linestyle='dashed', linewidth=2)
ax.axvline(median_value, color='red', linestyle='dashed', linewidth=2)

ax.legend()
```

```python
np.mean(np.abs(new_df['AmountUSD']- mean_value))
```

```python
new_df.fillna({'AmountUSD': median_value, inplace=True})
```

```python
data=new_df.groupby('ClientTier')['ClientTier'].count()
sns.barplot(data)
```
