```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataframe= pd.read_csv('new_clean.csv')
```

```python
df= dataframe.copy()
```

```python
df.head(50)
```

```python
df.info()
```

```python
df['date']=pd.to_datetime(df['date'], format= "%Y-%m-%d")
```

```python
df.info()
```

```python
null_date= df[df['sales'].isna()]['date'].values
null_date
```

```python
df['interplote_data']= df['sales'].interpolate(method='linear')
df['qudratic_intrp']= df['sales'].interpolate(method='quadratic')
df['cubic_intep']= df['sales'].interpolate(method='cubicspline')
df
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['date'], df['interplote_data'], '-', label= 'linear',color= 'red')
plt.plot(df['date'], df['qudratic_intrp'],'-', label= 'quadratic', color='blue')
plt.plot(df['date'], df['cubic_intep'],'-', label= 'cubic', color='violet')
plt.plot(df['date'], df['sales'], 'o', label= 'sales', markeredgewidth=1, markeredgecolor='black')


for line in null_date:
    plt.axvline(line, linestyle='dashed', color='black')
```

```python
err= df[['date','sales']].copy()
err['dummy']= err['sales']
err.loc[(err.index + 1) % 5 == 0, "dummy"] = np.nan
```

```python
err['linear']= err['dummy'].interpolate(method='linear')
err['quad']= err['dummy'].interpolate(method='quadratic')
err['cubic']= err['dummy'].interpolate(method='cubicspline')
err
```
